- **Control de acceso a la red**: Esto incluye implementar mecanismos que autentiquen a los usuarios y ordenadores antes de permitirles el acceso a la red. - **CID**: Confidencialidad, integridad y disponibilidad- es otro pilar importante de la seguridad por diseo para la arquitectura de redes. La confidencialidad consiste en mantener los datos de la organizacin privados y secretos. La integridad es la garanta de que los datos no han sido manipulados ni alterados. Esta separacin de roles es bastante flexible, pero su mayor activo es que se puede modificar cualquiera de las capas sin tener incidencia en el resto de estas. Esta peculiaridad aporta al modelo versatilidad y adaptabilidad en entornos distribuidos que evolucionan a gran velocidad.

En las arquitecturas de red, el modelo de tres capas organiza los dispositivos de red en tres niveles diferentes: la capa de acceso, la capa de distribucin y la capa ncleo. Las redes comerciales suelen usar esta arquitectura para mejorar la gestin, la escalabilidad y el rendimiento.

1. La capa de acceso est ms cerca de los usuarios finales. Los dispositivos en esta capa, que suelen ser interruptores de capa 2, permiten la conexin directa a dispositivos finales como computadoras y telfonos. La capa de acceso gestiona tareas importantes como la segmentacin de VLAN, la seguridad de puertos y el PoE (Power over Ethernet) para dispositivos como telfonos VoIP y puntos de acceso inalmbricos.
2. La capa de distribucin conecta la capa de acceso con la capa ncleo. Antes de transferir el trfico a la capa ncleo, su funcin principal es agregar el trfico de mltiples interruptores de acceso. Adems, implementa polticas de seguridad, balanceo de carga y enrutamiento entre VLANs. Los interruptores de capa 3 se utilizan con frecuencia en esta capa, lo que permite un mayor control y gestin del trfico.
3. La capa central, tambin conocida como la columna vertebral de la red, administra el trfico de alto rendimiento entre los diferentes segmentos de la red. La capa ncleo debe ser altamente redundante y eficiente y diseada para maximizar la velocidad de conmutacin de paquetes y reducir la latencia. Aqu se utilizan interruptores de alta capacidad y velocidad, que pueden manejar grandes cantidades de datos y crear conexiones rpidas entre los diversos mdulos de distribucin.

La estructura modular proporcionada por el diseo de tres capas facilita la expansin y el mantenimiento de la red. La arquitectura de ncleo colapsado, que combina las funciones de las capas de distribucin y el ncleo para reducir costos y simplificar el diseo, mientras se mantiene un nivel adecuado de rendimiento y redundancia, es una alternativa a este modelo en entornos ms pequeos.

## Complejidad Temporal de la funcin de bsqueda de caminos
<a id="Complejidad-Temporal-de-la-funcim-de-bsqueda-de-caminos"></a>

La funcin encontrar_caminos busca todos los caminos entre dos nodos en un grafo, considerando una restriccin de nivel para los nodos vecinos. La complejidad temporal de esta funcin puede ser analizada considerando los siguientes factores:

### Estructura del Grafo

El grafo podemos dividirlo en tres elementos esenciales para el calculo de la complejidad, N como el nmero de nodos del grafo, E como el nmero de aristas del grafo y D como el grado mximo de cualquier nodo del grafo.

### Recursividad

La funcin explora todos los caminos posibles desde el nodo actual hasta el nodo destino. En el peor de los casos, la profundidad de la recursin puede ser el nmero total de nodos $N$. A la hora de realizar la bsqueda de vecinos la funcin explora todos los vecinos del nodo actual. Si consideramos que, en el peor de los casos, cada nodo puede tener hasta $d$ vecinos, la funcin realiza $d$ llamadas recursivas por cada nodo.

El nmero de caminos potenciales entre dos nodos puede ser exponencial en el tamao de un grafo no dirigido y sin restricciones. En el peor de los casos, la funcin podra explorar todos los caminos disponibles, lo cual tiene una complejidad exponencial.

### Complejidad Temporal

Para determinar la complejidad temporal exacta de la funcin, consideramos los factores antes mencionados. En el peor de los casos:

1. Nmero de Caminos: La funcin tiene la capacidad de explorar un nmero exponencial de caminos. En el caso ms extremo, la funcin podra tener que explorar hasta $O(d^N)$ rutas, con $d$ vecinos para cada nodo y $N$ niveles de profundidad.
2. Verificacin de Niveles: Para cada vecino, la funcin verifica una condicin de nivel, que tiene una complejidad constante $O(1)$.

Por lo tanto, la complejidad temporal en el peor de los casos es: $O(d^N)$

### Verificacin de Niveles en la Funcin

Cada vez que se verifica si un vecino cumple con la condicin de nivel, se realizan las siguientes operaciones:

- Acceso al diccionario de nodos para obtener los atributos de un nodo especfico.
- Bsqueda de un atributo dentro del diccionario del nodo.
- Comparacin del nivel actual con el nivel requerido.

El proceso completo de verificacin de nivel para un nodo vecino es una operacin de tiempo constante porque todas estas operaciones (acceso al diccionario, bsqueda de clave en el diccionario y comparacin) tienen una complejidad $O(1)$.

## Recursividad vs Iteracin

### Recursividad

Lo primero que se nos viene a la mente a la hora de pensar de hacer algn cdigo con recursividad se trata de la pila de llamadas del sistema, esta nos permite guardar el estado actual de la exploracin y cada llamada guarda su estado en la pila mientras explora ms profundamente. A la hora de crear la funcin recursiva iba a ser un cdigo ms simple y fcil de explicar ya que nuestro problema se puede dividir en subproblemas de forma fcil. El nico problema que se nos viene a la mente de forma sencilla es que este mtodo esta limitado por la profundidad de la pila, esto nos puede dar problemas para grafos muy grandes o con caminos muy largos.

### Iteracin

Un enfoque iterativo puede ser ms difcil ya que es necesario el uso manual de la pila de llamadas para simular la recursin, pero esto nos puede dar cierta ventaja si queremos optimizar el uso de la memoria para asi evitar desbordamiento de la pila o estructura que estemos usando. Tenemos menos riesgo ya que como no tenemos un lmite en lo grande que es la pila de llamadas, podemos operar sobre cualquier tipo de grafo dando igual como de grande sea.

Tras optar por la opcin de un enfoque recursivo se le dio al algoritmo un enfoque nuevo donde habra que aadirle a cada nodo un atributo de nivel para por ver si dicho tipo de nodo dentro de la red puede dejar pasar paquetes de nivel X. Trabajaremos sobre el nivel 2, 3 y 4 de la capa OSI para poder determinar los distintos caminos del nivel X.
